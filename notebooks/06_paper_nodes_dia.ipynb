```python
import pandas as pd

filtered_df = pd.read_csv("../outputs/filtered_papers.csv")
metadata_df = pd.read_csv("../outputs/openalex_metadata.csv")

print("Filtered papers:", len(filtered_df))
print("Metadata rows:", len(metadata_df))
```

```python
merged_df = filtered_df.merge(
    metadata_df,
    on="title",
    how="left"
)

print("After merge:", len(merged_df))
print("Missing metadata:", merged_df["openalex_id"].isna().sum())
```

```python
paper_nodes = merged_df[[
    "global_paper_id",
    "title",
    "publication_year",
    "domain",
    "split",
    "openalex_id"
]].copy()

paper_nodes = paper_nodes.rename(columns={
    "global_paper_id": "node_id",
    "publication_year": "year"
})

paper_nodes["node_type"] = "Paper"

paper_nodes = paper_nodes[[
    "node_id",
    "node_type",
    "title",
    "year",
    "domain",
    "split",
    "openalex_id"
]]

paper_nodes.to_csv("../outputs/paper_nodes_dia.csv", index=False)

print("Saved paper_nodes.csv")
```
